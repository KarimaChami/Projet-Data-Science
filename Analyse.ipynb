```python
import pandas as pd 
data = pd.read_csv("data-68e11476082f9096032105.csv")
data["TotalCharges"] = pd.to_numeric(data["TotalCharges"],errors="coerce")
data["TotalCharges"] = data["TotalCharges"].fillna(data["TotalCharges"].mean())
# indique Ã  pandas que si une valeur ne peut pas Ãªtre convertie doit Ãªtre remplacÃ©e par NaN
data.dtypes
print(data.isnull().sum())
```

```python

import matplotlib.pylab as plt
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder

label_oncoder  =  LabelEncoder()
data["Churn"] = label_oncoder.fit_transform(data["Churn"])

cat_cols = data.select_dtypes(include=['object']).columns
for col in cat_cols:
 if col != 'customerID':
   sns.countplot(data=data,x=data[col],hue='Churn')
   plt.show()
  #  plt.figure(figsize=(12,4))
   
print(data['Churn'].value_counts(normalize=True)) #le pourcentage de clients dÃ©sabonnÃ©s vs fidÃ¨les.


```
